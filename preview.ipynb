```python
# keep the ID1, ID2, TIETYPE, RACE1, RACE2, SEX1, SEX2, ORIENT1, 
# ORIENT2, BEHAV1, BEHAV2, AGE1, AGE2, DEALER1, DEALER2, HWIFE1, HWIFE2,
# DISABLE1, DISABLE2, UnEMP1, UnEMP2, STREETS,YEAR, STDYYEAR,
# EDUC1, EDUC2, HIV1, HIV2, FREQ, YOB1, YOB2 columns in df
# Define the list of columns to keep
columns_to_keep = ['ID1', 'ID2', 'TIETYPE', 'RACE1', 'RACE2', 'SEX1', 'SEX2', 'ORIENT1', 'ORIENT2',
                   'BEHAV1', 'BEHAV2', 'AGE1', 'AGE2', 'DEALER1', 'DEALER2', 'HWIFE1', 'HWIFE2',
                   'DISABLE1', 'DISABLE2', 'UNEMP1', 'UNEMP2', 'STREETS1', 'STREETS2','YEAR', 'STDYYEAR',
                   'EDUC1', 'EDUC2', 'HIV1', 'HIV2', 'FREQ', 'YOB1', 'YOB2','FRSTYR']

# Use loc to select only the desired columns
df_ini = df.loc[:, columns_to_keep]
```

```python
# Set the Pandas display options to show all columns
pd.set_option('display.max_columns', None)
display(df_ini.head(), df_ini.shape)
```

```python
# drop the ' ' values in all the columns in df_ini
df_filtered = df_ini.replace(' ', np.nan)

# drop the rows that have any missing values in df_ini
df_filtered = df_filtered.dropna()
df_filtered.shape
```

```python
# change all the data type in df_filtered to numeric
df_filtered = df_filtered.apply(pd.to_numeric)

# check if there is aby non-numeric columns in df_filtered
df_filtered.select_dtypes(exclude=[np.number]).shape
```

```python
# create a new column called weight in df_filtered
# the weight is counting the number of rows in each group of ID1 and ID2
# create a new df called df_weighted by grouping the df_filtered by ID1 and ID2
# and summing the weight column
df_weighted = df_filtered.groupby(['ID1', 'ID2']).size().reset_index(name='Weight')
display(df_weighted.head(), df_weighted.shape)
```

```python
# check if there is any duplicated rows on ID1 and ID2 in df_weighted
df_weighted.duplicated(subset=['ID1','ID2']).sum()
```

```python
# save the df_weighted as a csv file
df_weighted.to_csv('edgelist.csv', index=False)
```
